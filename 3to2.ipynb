```python
import pandas as pd

# Load your Excel file into a DataFrame
file_path = r"C:\Users\Msc 2\Downloads\manipulated_data2.xlsx"
df = pd.read_excel(file_path)

# Check which rows have no ones
no_ones_rows = df[(df == 1).sum(axis=1) == 0]

# Print the rows with no ones
print(no_ones_rows)
```

```python
df = df.drop(index=no_ones_rows.index)

# Print the updated DataFrame
print(df)
```

```python
# Specify the path and filename for the updated Excel file
output_file_path = r"C:\Users\Msc 2\Downloads\manipulated_data3.xlsx"

# Save the updated DataFrame to an Excel file
df.to_excel(output_file_path, index=False)
```
